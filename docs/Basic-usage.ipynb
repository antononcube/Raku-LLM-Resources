## Resource graphs

Translation by fallback LLM-graph:

```raku
my $spec = q:to/END/;
make a brand new recommender with the data @dsData;
apply LSI functions IDF, None, Cosine; 
recommend by profile for passengerSex:male, and passengerClass:1st;
join across with @dsData on "id";
echo the pipeline value;
END

my $llm-evaluator = llm-evaluator(llm-configuration('Ollama', model => 'gemma3:4b'));
my $gBestCode = llm-resource-graph('code-generation-by-fallback', input => {:$spec, lang => 'Raku', workflow-name => 'SMRMon', :split}, :$llm-evaluator);
```

The translation result:

```raku
$gBestCode.nodes<code><result>
```

The corresponding graph plot:

```raku
#% html
$gBestCode.dot(:6size, :10vertex-font-size, vertex-width => 1, theme => 'default'):svg
```

## CLI

Using a smaller LLM in order to get results faster:

```raku
shell "llm-text-summarization $url --conf=ollama::gpt-oss:20b --with-title=Whatever -o=url-summary.md";
```

The first few thousand characters of the summary:

```raku
#% markdown
slurp('./url-summary.md').substr(^2000)
```
